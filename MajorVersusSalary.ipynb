```python
totalDiffs = clean_moneyStats["Mid-Career 90th Percentile Salary"] - clean_moneyStats["Mid-Career 10th Percentile Salary"]
clean_moneyStats.insert(1, "Start and Later Salary Diffs", totalDiffs)
# inserting into the existing dataframe
# we want the column to be the 2nd column, so position 1
clean_moneyStats.head()
```

```python
lowRisk = clean_moneyStats.sort_values("Start and Later Salary Diffs")
# to see which values in that column are the smallest
# this strictly just sorts them from smallest to largest

lowRisk[["Undergraduate Major", "Start and Later Salary Diffs"]].head()
# looks at first 5 rows
# to look at multiple columns, you need double brackets to pass them as list
# to indexing brackets
```

```python
# we have 3 groups: STEM, HASS, business
clean_moneyStats.groupby("Group").count()
# this organizes our data by group
# groupby() allows you to group data based on specified fields
# in this case, based on the frequency of "Group" column values (the 3 mentioned)
```

```python
pd.options.display.float_format = "{:,.2f}".format
# for {:,.2f} the : is the placeholder for a number to be formatted
# the , is what indicates we want commas to separate values by thousands
# the .2f is what rounds the number 2 decimal places and makes sure it is a float
clean_moneyStats.groupby("Group").mean(numeric_only = True)
# numeric_only = True ensures that only the numeric columns are considered for means
```
